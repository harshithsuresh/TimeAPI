```python
import json
from datetime import datetime

f = open('input1.json')

data = json.load(f)
start_time = datetime.strptime("2021-01-28T07:30:00Z","%Y-%m-%dT%H:%M:%SZ")
end_time = datetime.strptime("2021-01-28T14:30:00Z","%Y-%m-%dT%H:%M:%SZ")

def in_between(now, start, end):
    if start <= end:
        return start <= now < end
    else: 
        return start <= now or now < end
    
def get_shift(hour):
    if hour>=6 and hour<14:
        return "shiftA"
    elif hour>=14 and hour<20:
        return "shiftB"
    else:
        return "shiftC"    
res={
"shiftA" :{ "production_A_count" :0, "production_B_count" :0},
"shiftB" :{ "production_A_count" :0, "production_B_count" :0},
"shiftC" :{ "production_A_count" :0, "production_B_count" :0},
}
for row in data:
    time=datetime.strptime(row['time'], "%Y-%m-%d %H:%M:%S")
    if in_between(time,start_time,end_time):
        shift=get_shift(time.hour)
        if row['production_A']==True:
            res[shift]['production_A_count']+=1
        if row['production_B']==True:
            res[shift]['production_B_count']+=1
print(res)            
        

    
```

```python
import json
from datetime import datetime

def convert(seconds):
    minutes, seconds = divmod(seconds, 60)
    hours, minutes = divmod(minutes, 60)
    return str(hours)+"h:"+str(minutes)+"m:"+str(seconds)+"s"

f = open('input2.json')
start_time = datetime.strptime("2021-01-28T08:30:00Z","%Y-%m-%dT%H:%M:%SZ")
end_time = datetime.strptime("2021-01-28T10:30:00Z","%Y-%m-%dT%H:%M:%SZ")

data = json.load(f)
downtime=0
runtime=0
res={}
for row in data:
    time=datetime.strptime(row['time'], "%Y-%m-%d %H:%M:%S")
    if in_between(time,start_time,end_time):
        if runtime>1021:
            downtime+=row['runtime']-1021
            runtime+=1021
        else:
            runtime+=row['runtime']
totalTime=(end_time-start_time).total_seconds()
utilisation=totalTime/(totalTime+downtime)*100
res["runtime"]=convert(runtime)
res["downtime"]=convert(downtime)
res["utilisation"]=round(utilisation,2)
```

```python
import json
from datetime import datetime


f = open('input3.json')
start_time = datetime.strptime("2021-01-28T18:30:00Z","%Y-%m-%dT%H:%M:%SZ")
end_time = datetime.strptime("2021-01-28T20:10:00Z","%Y-%m-%dT%H:%M:%SZ")

data = json.load(f)
res=[]
for row in data:
    time=datetime.strptime(row['time'], "%Y-%m-%d %H:%M:%S")
    if in_between(time,start_time,end_time):
        if row['state']==True:
            res.append({'id':int((row['id'])[-3:]),'belt1':0,'belt2':row['belt2']})
        else:
            res.append({'id':int((row['id'])[-3:]),'belt1':row['belt1'],'belt2':0})
            
            
        
```
